# This Notebook Shows the Contents of the Exhange Rate Dataset
- Dataset can be found here: https://www.kaggle.com/robikscube/exhange-rates-to-usd-from-imforg-updated-daily/

```python
import pandas as pd
from itertools import cycle
import matplotlib.pylab as plt

plt.style.use("seaborn-whitegrid")
color_cycle = cycle(plt.rcParams["axes.prop_cycle"].by_key()["color"])
```

```python
to_usd_df = pd.read_csv('../input/exhange-rates-to-usd-from-imforg-updated-daily/exchange_rate_to_usd.csv')
to_usd_df = pd.read_parquet('../input/exhange-rates-to-usd-from-imforg-updated-daily/exchange_rate_to_usd.parquet')
```

```python
for exrate in to_usd_df.columns:
    if exrate == 'date':
        continue
    to_usd_df[exrate] \
        .plot(figsize=(15, 5), title=exrate, color=next(color_cycle))
    plt.show()
```
